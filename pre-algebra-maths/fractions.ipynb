# Fractions

The top number is called the numerator & the bottom number is called the denominator.

### Lowest Common Multiple
When 2 fractions need a common denominator to be able to perform
arithmetic on, then finding the lowest natural number that both fractions can use as their denominator


### Cancelling
If you work out the equivalent denominators of 2 fractions  by dividing or multiplying then its called cancelling.
