# Number

### Order Of Operation

Carry out the calculation inside brackets first.

1. If there are no brackets then - **multiplication & division** should be carried out first.

e.g `3 + 4 * 2` should be carried out as `4 * 2 = 8` then `8 + 3 = 11`.

# Multiples, Factors & Primes
### Multiples
Basically the times table of each number -
1*8, 2*8, 3*8

### Factors
Numbers that divide exactly into a number -
For 18 this would nbe - 1, 2, 3, 6.

- `1` is a factor of every number.
- Every number is a factor of itself.

### Primes
A number greater than 1 that has exactly 2 factors, `1` & itself.
