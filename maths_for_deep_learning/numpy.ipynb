# Defining Arrays
We use numpy arrays as they are much faster than Python lists.

Numpy under the hood is implemented in C, so it supports the same set of data types that C supports.

```python
import numpy as np
```

```python
a = np.array([1, 2, 3, 4])
a
```

```python
a.size
```

```python
# shape shows that 'a' is a single row with 4 columns.
a.shape
```

```python
a.dtype
```

```python
# Multi Dimensional array example
row_a = [1, 2, 3, 4]
row_b = [5, 6, 7, 8]
b = np.array([row_a, row_b])
b.shape # (2, 4) - 2 rows, 4 columns
```
