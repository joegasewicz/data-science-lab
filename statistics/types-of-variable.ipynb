# Types of Variable
In statistics variables are applied as:
- Category
- Available stats/ visualizations

## Category Types
- Numeric
- Categoric


## 5 Types Of Variables

### Interval
- Category: Numeric.
- Type: Interval
- Description: Numeric scale; Arbitrary precision.
- Example: Temperature in Celsius.

### Ratio
- Category: Numeric.
- Type: Ratio
- Description: Interval with meaningful 0.
- Example: Money in your bank account.

### Discrete
- Category: Numeric.
- Type: Discrete
- Description: Integer with no arbitrary precision.
- Example: Population (you can't have half a person).

### Ordinal
- Category: Categoric.
- Type: Ordinal
- Description: Sortable, discrete & without scale. (e.g. we can sort it because BA is more than MA but we don't know how much more better it is).
- Example: Level of education (BA, MA, PhD etc).

### Nominal
- Category: Numeric.
- Type: Nominal
- Description: Non-sortable, discrete.
- Example: Gender (female, male) - we can't say which is better.

### Examples
- Yearly budget deficit/surplus: Ratio (does have a meaningful 0).
- Height in inches: Ratio (does have a meaningful zero).
- Top 3 fave films: Ordinal (it's sortable but now way to quantify how big the difference is between values).
- Intelligence score (IQ): Interval (There's no meaningful zero).
- place in a race: Ordinal (Can't quantify difference in a race with an ordinal scale).
- Time on a clock: Interval (There's no meaningful zero)
- Profession: Nominal (we can sort them but we can't order or quantify them).
- Iphones sold: Discrete (can't sell half an iphone).
